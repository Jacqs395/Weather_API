```python
import pandas as pd
from pandas import Series, DataFrame

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/sa18/Desktop/weather_api/new.csv")

df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

df = df.dropna()

df["time"] = pd.to_datetime(df["time"])

df = df.set_index("time")

```

```python
## median temperature for all available months & years
monthly_median_temperature = df.resample("ME")["temperature"].median()
print(monthly_median_temperature)
monthly_median_temperature.to_csv("monthly_median_temp.csv")

yearly_median_temperature = df.resample("YE")["temperature"].median()
print(yearly_median_temperature)
yearly_median_temperature.to_csv("yearly_median_temp.csv")
```

```python
## median humidity for all available months & years

monthly_median_humidity = df.resample("ME")["humidity"].median()
print(monthly_median_humidity)
monthly_median_humidity.to_csv("monthly_median_humidity.csv")

yearly_median_humidity = df.resample("YE")["humidity"].median()
print(yearly_median_humidity)
yearly_median_humidity.to_csv("yearly_median_humidity.csv")
```

```python
## median precipitation for all available months & years

monthly_median_precipitation = df.resample("ME")["precipitation"].median()
print(monthly_median_precipitation)
monthly_median_precipitation.to_csv("monthly_median_precipitation.csv")

yearly_median_precipitation = df.resample("YE")["precipitation"].median()
print(yearly_median_precipitation)
yearly_median_precipitation.to_csv("yearly_median_precipitation.csv")
```

```python
## median surface pressure for all available months & years

monthly_median_surface_pressure = df.resample("ME")["surface_pressure"].median()
print(monthly_median_surface_pressure)
monthly_median_surface_pressure.to_csv("monthly_median_surface_pressure.csv")

yearly_median_surface_pressure = df.resample("YE")["surface_pressure"].median()
print(yearly_median_surface_pressure)
yearly_median_surface_pressure.to_csv("yearly_median_surface_pressure.csv")
```

```python
## yearly median along with harvest data

df1 = pd.read_csv("/Users/sa18/Desktop/weather_api/yearly_median_temp.csv")
df2 = pd.read_csv("/Users/sa18/Desktop/weather_api/minas_data.csv")

# convert the "time" column in df1 to datetime and extract the year
df1['time'] = pd.to_datetime(df1['time'], format='%Y-%m-%d')

# extract year from "time"
df1['year'] = df1['time'].dt.year

# make 'year' column in df2 an integer
df2['year'] = df2['year'].astype(int)

# merge df1 and df2 according to year
yearly_median_temp_harvest_data = pd.merge(df1, df2, on='year', how='inner')

print("yearly_median_temp_harvest_data: ")
print(yearly_median_temp_harvest_data.head())

yearly_median_temp_harvest_data.to_csv("/Users/sa18/Desktop/weather_api/yearly_median_temp_harvest_data.csv", index=False)
```
