```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/sa18/Desktop/weather_api/data/csv/br_final.csv")
```

```python
#How have harvest outcomes changed for Minas Gerais throughout the years?  

minas_data = df[df['subdivision'] == 'Minas Gerais']

# coffee harvest for Minas Gerais over time
plt.figure(figsize=(12, 4))
sns.lineplot(x='year', y='million_60kgs_bag', data=minas_data, marker='o', color='green')

# plot
plt.title("Coffee Harvest in Minas Gerais (Million 60kg Bags) Through the Years", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Million 60kg Bags", fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

How have harvest outcomes changed for Minas Gerais throughout the years? From the plot above, it looks like coffee harvest in Minas Gerais fluctuated from 2005-2020. Overall, however, the harvest has increased since 2005. Somewhere around ~2008, Minas Gerais' harvest increased and maintained an uptick, with slight dips. In the years ~2013-2014, there was a steady increase in production. However, around 2022, production dropped dramatically, before a slight gain around 2023.
